```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("suiciderates.csv")
```

```python
df = df.drop(df.columns[[0,1,3,12]], axis = 1)
```

```python
df.isnull().sum()
#Checks for any null value within the data set. Keep ESTIMATE nulls
```

```python
male_all_age = df.iloc[42:84]

male_all_age.groupby('YEAR')['ESTIMATE'].mean().plot(kind='bar', color='orange')
plt.title('Suicide rate by Sex')
plt.xlabel('Male')
plt.ylabel('Suicide rate by 100,000 Resident Population')
```

```python
female_all_age = df.iloc[85:126]

female_all_age.groupby('YEAR')['ESTIMATE'].mean().plot(kind='bar', color='red')
plt.title('Suicide rate by Sex')
plt.xlabel('Female')
plt.ylabel('Suicide rate by 100,000 Resident Population')
```

```python
m10_14 = df[1526:1568]
m15_19 = df[1610:1652]
m20_24 = df[1652:1694]
m25_34 = df[1736:1778]
#Cut off at 25-34 since it won't be considered young anymore
f10_14 = df[2114:2156]
f15_19 = df[2198:2240]
f20_24 = df[2240:2282]
f25_34 = df[2324:2366]
```

```python
plt.plot(m10_14['YEAR'],m10_14['ESTIMATE'], label = "10-14 years old",
         color = "r")
plt.plot(m15_19['YEAR'],m15_19['ESTIMATE'], label = "15-19 years old",
         color = "b")
plt.plot(m20_24['YEAR'],m20_24['ESTIMATE'], label = "20-24 years old",
         color = "g")
plt.plot(m25_34['YEAR'],m25_34['ESTIMATE'], label = "25-34 years old",
         color = "k")
plt.legend(bbox_to_anchor = (1.02, 1), loc = 'upper left', borderaxespad = 0)
plt.xlabel("Year")
plt.ylabel("Death By 100,000 Resident Population")
plt.title("Male Suicide Through The Decades")
```

```python
plt.plot(f10_14['YEAR'],f10_14['ESTIMATE'], label = "Female 10-14 years old",
         color = "r")
plt.plot(f15_19['YEAR'],f15_19['ESTIMATE'], label = "Female 15-19 years old",
         color = "b")
plt.plot(f20_24['YEAR'],f20_24['ESTIMATE'], label = "Female 20-24 years old",
         color = "g")
plt.plot(f25_34['YEAR'],f25_34['ESTIMATE'], label = "Female 25-34 years old",
         color = "k")
plt.legend(bbox_to_anchor = (1.02, 1), loc = 'upper left', borderaxespad = 0)
plt.xlabel("Year")
plt.ylabel("Suicide By 100,000 Resident Population")
plt.title("Female Suicide Through The Decades")
```

```python
plt.plot(m10_14['YEAR'],m10_14['ESTIMATE'], label = "Male 10-14 years old",
         color = "r")
plt.plot(m15_19['YEAR'],m15_19['ESTIMATE'], label = "Male 15-19 years old",
         color = "b")
plt.plot(m20_24['YEAR'],m20_24['ESTIMATE'], label = "Male 20-24 years old",
         color = "g")
plt.plot(m25_34['YEAR'],m25_34['ESTIMATE'], label = "Male 25-34 years old",
         color = "k")
plt.plot(f10_14['YEAR'],f10_14['ESTIMATE'], label = "Female 10-14 years old",
         linestyle="--",color = "r")
plt.plot(f15_19['YEAR'],f15_19['ESTIMATE'], label = "Female 15-19 years old",
         linestyle="--", color = "b")
plt.plot(f20_24['YEAR'],f20_24['ESTIMATE'], label = "Female 20-24 years old",
         linestyle="--", color = "g")
plt.plot(f25_34['YEAR'],f25_34['ESTIMATE'], label = "Female 25-34 years old",
         linestyle="--", color = "k")
plt.legend(bbox_to_anchor = (1.02, 1), loc = 'upper left', borderaxespad = 0)
plt.xlabel("Year")
plt.ylabel("Suicide By 100,000 Resident Population")
plt.title("Male vs. Female Suicide Comparison")
```
